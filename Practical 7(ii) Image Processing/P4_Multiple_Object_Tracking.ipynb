```python
import cv2                          #complete this
import sys                         #complete this
from random import randint

tracker_types = ['BOOSTING', 'MIL', 'KCF', 'TLD', 'MEDIANFLOW', 'MOSSE', 'CSRT']             #declare detection type

def create_tracker_by_name(tracker_type):
    if tracker_type == tracker_types[0]:
        return cv2.legacy.TrackerBoosting_create()
    elif tracker_type == tracker_types[1]:
        return cv2.legacy.TrackerMIL_create()
    elif tracker_type == tracker_types[2]:
        return cv2.legacy.TrackerKCF_create()
    elif tracker_type == tracker_types[3]:
        return cv2.legacy.TrackerTLD_create()
    elif tracker_type == tracker_types[4]:
        return cv2.legacy.TrackerMedianFlow_create()
    elif tracker_type == tracker_types[5]:
        return cv2.legacy.TrackerMOSSE_create()
    elif tracker_type == tracker_types[6]:
        return cv2.legacy.TrackerCSRT_create()
    else:
        print('Invalid tracker name. Available trackers:')
        for t in tracker_types:
            print(t)
        return None

video = cv2.VideoCapture('Videos/race.mp4')                           #complete this
if not video.isOpened():
    print('Error while loading the video!')
    sys.exit()

ok, frame = video.read()
if not ok:
    print('Error while loading the first frame!')
    sys.exit()

bboxes = []
colors = []

while True:
    bbox = cv2.selectROI('MultiTracker', frame, False)
    if bbox[2] != 0 and bbox[3] != 0:  
        bboxes.append(bbox)
        colors.append((randint(0, 255), randint(0, 255), randint(0, 255)))
    else:
        print("No valid ROI selected. Try again.")

    print('Press SPACE to select the next object, Q to start tracking, or ESC to exit.')
    k = cv2.waitKey(0)
    print(f"Key pressed: {k}")  
    if k == ord('q') or k == ord('Q'):  
        break
    elif k == 27:  # ESC
        cv2.destroyAllWindows()
        sys.exit()

print(bboxes)
print(colors)

tracker_type = 'CSRT'                                       #complete this
multi_tracker = cv2.legacy.MultiTracker_create()
for bbox in bboxes:
    tracker = create_tracker_by_name(tracker_type)
    if tracker:
        multi_tracker.add(tracker, frame, bbox)

while video.isOpened():
    ok, frame = video.read()
    if not ok:
        break

    ok, boxes = multi_tracker.update(frame)
    for i, new_box in enumerate(boxes):
        (x, y, w, h) = [int(v) for v in new_box]
        cv2.rectangle(frame, (x, y), (x + w, y + h), colors[i], 2)

    cv2.imshow('MultiTracker', frame)
    if cv2.waitKey(1) & 0xFF == 27:  
        break
        
cv2.waitKey(0)
video.release()  # Release the video capture object
cv2.destroyAllWindows()  # Close all OpenCV windows
```
