# Object detection using MobileNetV3 (image)

```python
import cv2                     #complete this
import matplotlib.pyplot as plt                     #complete this 
```

```python
config_file='ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'                #complete this
frozen_model='frozen_inference_graph.pb'               #complete this
```

```python
model = cv2.dnn_DetectionModel(frozen_model, config_file)                     #complete this
```

```python
classLabels = []
file_name = 'labels.txt'
with open (file_name,'rt')as fpt:
    classlabels = fpt.read().rstrip('\n').split('\n')
```

```python
print (classlabels)
print (len(classlabels))
```

```python
model.setInputSize(320,320)             
model.setInputScale(1.0/127.5)            
model.setInputMean((127.5,127.5,127.5))        
model.setInputSwapRB(True)
```

```python
img=cv2.imread('car2.jpg')                                      #complete this
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
classIds, confs, bbox=model.detect(img, confThreshold=0.7)          #complete this
```

'confThreshold': This is the confidence threshold used to filter out weak detections. Only detections with confidence scores greater than this threshold are considered. In this case, confThreshold=0.7 means that only detections with confidence scores above 70% will be included.

```python
print(bbox)
```

```python
print(confs) # confidence lvl
```

```python
print(classIds)
```

```python
font_scale = 5              
font = cv2.FONT_HERSHEY_PLAIN                     #complete this
for ClassInd, conf, boxes in zip(classIds.flatten(), confs.flatten(), bbox):  
    cv2.rectangle(img, boxes, (255, 0, 0), 4)   
    cv2.putText(img, classlabels[ClassInd-1], (boxes[0]-10, boxes[1]-40), font, fontScale=font_scale, color=(0, 255, 255), thickness=5)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
